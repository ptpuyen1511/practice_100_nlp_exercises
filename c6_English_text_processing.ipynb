# 50. Tách câu (Sentence segmentation)
Sử dụng patterns (. or ; or : or ? or !) -> ký tự space -> chữ cái tiếng Anh viết hoa (captial letter) để tách các câu trong văn bản. Đầu vào là văn bản nlp.txt, in ra mỗi câu trong văn bản trên 1 dòng.

```python
import re

with open('nlp.txt', 'r') as f:
    text = f.readlines()
    
list_sentences = []

for line in text:
    if line == '\n':
        continue
        
    sentences = re.split(r'(?<=[^A-Z].[.;:?!]) +(?=[A-Z])', line.strip())
    
    for sentence in sentences:
        print(sentence)
        list_sentences.append(sentence)
```

# 51. Tách từ
Coi ký tự trắng (space) là ký tự phân tách các từ. Lấy đầu ra của bài 50 làm đầu vào, trích xuất các từ trong các câu và in ra theo định dạng: mỗi dòng 1 từ. In ra dòng trắng để đánh dấu kết thúc câu.

```python
list_sentence_splited_word = []
for sentence in list_sentences:
    sentence = re.sub(r'(?<=[^A-Z].)[.;:?!](?=[ $]*)', '', sentence) # Remove end of sentence
    sentence = re.sub(r'([\(\)"])', ' \g<1> ', sentence) # add space for ( )
    words = sentence.replace(',', '').split() # remove , --> split
    
    list_sentence_splited_word.append(words)
    for word in words:
        print(word)
        
    print()
```

# 52. Stemming
Đầu vào là đầu ra của bài tập 51, áp dụng thuật toán Porter stemming để lấy ra gốc của các từ (stem). In ra các từ và stem cách nhau bởi ký tự tab. Nếu bạn sử dụng Python, bạn có thể sử dụng module stemming.

```python
! pip install stemming
```

```python
from stemming.porter2 import stem

for sentence_splited_word in list_sentence_splited_word:
    for word in sentence_splited_word:
        print(word + '\t' + stem(word))
    print()
```

# 53. Tokenization
Sử dụng tool Stanford Core NLP để phân tích văn bản đầu vào và lấy ra output theo định dạng XML. Sau đó đọc vào đầu ra XML và trích xuất ra các token (word) theo định dạng mỗi word trên 1 dòng.

```python
!pip install stanfordnlp
```

```python
import stanfordnlp
import re
```

```python
nlp = stanfordnlp.Pipeline(processors='tokenize', lang='en')
```

```python
# Create input for stanford nlp
# Read data
with open('nlp.txt', 'r') as f:
    text = f.readlines()

# Strip \n
text = [s.strip() for s in text if s.strip() != '']

# Add punctuation .
text = [s + '.' if s[-1] != '.' else s for s in text]

# Convert to 1 string
text_one_str = ' '.join(text)
```

```python
doc = nlp(text_one_str)
```

```python
for i, sentence in enumerate(doc.sentences):
    print(f"----- Sentence {i+1} tokens -----")
    print(*[f"{token.text}" for token in sentence.tokens], sep='\n')
```
