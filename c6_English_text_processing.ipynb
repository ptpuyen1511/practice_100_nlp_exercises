# 50. Tách câu (Sentence segmentation)
Sử dụng patterns (. or ; or : or ? or !) -> ký tự space -> chữ cái tiếng Anh viết hoa (captial letter) để tách các câu trong văn bản. Đầu vào là văn bản nlp.txt, in ra mỗi câu trong văn bản trên 1 dòng.

```python
import re

with open('nlp.txt', 'r') as f:
    text = f.readlines()
    
list_sentences = []

for line in text:
    if line == '\n':
        continue
        
    sentences = re.split(r'(?<=[^A-Z].[.;:?!]) +(?=[A-Z])', line.strip())
    
    for sentence in sentences:
        print(sentence)
        list_sentences.append(sentence)
```

# 51. Tách từ
Coi ký tự trắng (space) là ký tự phân tách các từ. Lấy đầu ra của bài 50 làm đầu vào, trích xuất các từ trong các câu và in ra theo định dạng: mỗi dòng 1 từ. In ra dòng trắng để đánh dấu kết thúc câu.

```python
list_sentence_splited_word = []
for sentence in list_sentences:
    sentence = re.sub(r'(?<=[^A-Z].)[.;:?!](?=[ $]*)', '', sentence) # Remove end of sentence
    sentence = re.sub(r'([\(\)"])', ' \g<1> ', sentence) # add space for ( )
    words = sentence.replace(',', '').split() # remove , --> split
    
    list_sentence_splited_word.append(words)
    for word in words:
        print(word)
        
    print()
```

# 52. Stemming
Đầu vào là đầu ra của bài tập 51, áp dụng thuật toán Porter stemming để lấy ra gốc của các từ (stem). In ra các từ và stem cách nhau bởi ký tự tab. Nếu bạn sử dụng Python, bạn có thể sử dụng module stemming.

# 56. Phân tích coreference
Dựa trên kết quả phân tích coreference của Stanford Core NLP, thay thế các mention bằng representative mention. Chú ý khi thay thế các mention bằng representative mention, lưu lại các mention gốc trong dấu ngoặc theo định dạng representative mention (mention).

```python
import xml.etree.cElementTree as et
import pandas as pd

tree = et.parse('nlp.txt.xml')
root = tree.getroot()
```

```python
# Get coref from xml file
coref_dict = {}

for node in root.iter('coreference'):
    root_coref = et.Element('root') # create empty xml node
    root_coref = node
    
    repr_text = ''
    for mention in root_coref.iter('mention'):
        if mention.get('representative') == "true":
            repr_text = mention.find('text').text
        else:
            coref_dict[(int(mention.find('sentence').text), int(mention.find('start').text), int(mention.find('end').text))] = repr_text
```

```python
coref_dict
```

```python
# Get sentences corresponds to coref
sentences = []
for node in root.iter('sentences'):
    root_sentences = et.Element('root')
    root_sentences = node
    
    for sentence in root_sentences.iter('sentence'):
        root_tokens = et.Element('root')
        root_tokens = sentence
        sentence = []
        for token in root_tokens.iter('token'):
            sentence.append(token.find('word').text)
            
        sentences.append(sentence)
```

```python
# Repalce mention by mention representative mention
for k,v in coref_dict.items():
    sentences[k[0]-1] = sentences[k[0]-1][:k[1]-1] + [v, '('] + sentences[k[0]-1][k[1]-1:k[2]-1] + [')'] + sentences[k[0]-1][k[2]-1:]
```

```python
# Join sentences after repalce mention by mention representative mention
sentences = [' '.join(s) for s in sentences]

# Replace -LRB- -RRB-
sentences = [s.replace('-LRB-', '(').replace('-RRB-', ')').replace('( ', '(').replace('`` ', '"').replace('\'\'', '"') for s in sentences]

# Replace " punctuation" --> "punctuation"
sentences = [re.sub(r' ([).,\"])', '\g<1>', s) for s in sentences]
```

# 57. Phân tích cấu trúc dependency
Từ kết quả phân tích dependency của Stanford Core NLP (collapsed-dependencies), visualize câu đầu vào bằng đồ thị có hướng. Khi visualize các dependency trees, có thể sử dụng ngôn ngữ DOT, hay sử dụng Graphviz. Nếu sử dụng Python, có thể hiển thị đồ thị có hướng bằng thư viện/package pydot.

```python
import xml.etree.cElementTree as et
import pandas as pd

tree = et.parse('nlp.txt.xml')
root = tree.getroot()
```

```python
list_sentences_dep = []

for node in root.iter('dependencies'):
    if node.get('type') != 'collapsed-dependencies':
        continue
        
    root_2 = et.Element('root')
    root_2 = node
    
    dep_dict = {}
    
    for dep in root_2.iter('dep'):
        rel = dep.get('type')
        obj1 = dep.find('governor').text + '-' + dep.find('governor').get('idx')
        obj2 = dep.find('dependent').text + '-' + dep.find('dependent').get('idx')
        
        dep_dict[(obj1, obj2)] = rel
        print(rel+'(' + obj1 + ', ' + obj2 + ')')
     
    list_sentences_dep.append(dep_dict)
    print('---------------------')
```

```python
# Visualize
from graphviz import Digraph

u = Digraph('unix', filename='sentence_5_dep_parse.gv',
            node_attr={'color': 'lightblue2', 'style': 'filled'})
u.attr(size='6,6')

# Display sentence 5
for k,v in list_sentences_dep[4].items():
    u.edge(k[0], k[1], label=v)
```

# 58. Trích xuất tuples
Từ kết quả phân tích dependency của Stanford Core NLP (collapsed-dependencies), trích xuất các bộ 3 [Subject, Predicate, Purpose] và in ra các bộ 3 này (các thành phần cách nhau bởi ký tự tab). Subject, Predicate, Purpose được xác định dựa vào các tiêu chuẩn sau:

* Predicate: Là word ở các có các node con (dependant) trong các dependency relations: nsubj, dobj
* Subject: Các node con (dependant) trong các quan hệ nsubj từ predicate
* Purpose: Các node con (dependant) trong các quan hệ dobj từ predicate

```python
import xml.etree.cElementTree as et
import pandas as pd

tree = et.parse('nlp.txt.xml')
root = tree.getroot()
```

```python
list_sentences_dep = []

for node in root.iter('dependencies'):
    if node.get('type') != 'collapsed-dependencies':
        continue
        
    root_2 = et.Element('root')
    root_2 = node
    
    dep_dict = {}
    
    for dep in root_2.iter('dep'):
        rel = dep.get('type')
        obj1 = dep.find('governor').text + '-' + dep.find('governor').get('idx')
        obj2 = dep.find('dependent').text + '-' + dep.find('dependent').get('idx')
        
        dep_dict[(obj1, obj2)] = rel
     
    list_sentences_dep.append(dep_dict)
```

```python
# Browse for sentence
for i,sent_dep in enumerate(list_sentences_dep):
    # Browse nsubj
    for k1,v1 in sent_dep.items():
        if v1 == 'nsubj':
            for k2,v2 in sent_dep.items():
                if v2 == 'dobj':
                    if k1[0] == k2[0]:
                        print((k1[1].split('-')[0], k1[0].split('-')[0], k2[1].split('-')[0]))
```
